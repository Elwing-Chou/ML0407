```python
def test(a, **kwargs):
    print(a)
    print(kwargs)
test(5, c=1, d=5, e=6)
```

```python
def test2(a, r=1, b=2):
    print(a * r + b)
params = {
    "r":5,
    "b":6
}
test2(5, **params)
```

```python
def test3(a, *args):
    print(a)
    print(args)
test3(5, 2, 1, 4, 3)
```

```python
def test3(a, b, c, d):
    print(a, b, c, d)
params = [2, 5, 3]
test3(10, *params)
```
